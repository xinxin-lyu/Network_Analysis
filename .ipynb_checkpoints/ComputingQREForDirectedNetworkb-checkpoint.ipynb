# Part 1: Preliminaries

```python
# CRRA utility function
def CRRA(x,gamma): #can use offset by xMin in case of negative values of x
    if gamma == 1:
        return np.log(x)
    else:
        return ((x)**(1-gamma) - 1)/(1-gamma)
```

```python
# Define the payoff, assuming that the utility depends solely on the payoff
# Estimate the prob of decising at turn 0, each subject has 11 strategies, at each node (A->E), wait or decide at turn 0
# The difference (compared to network (a)) here is that when a subject is placed at A, his strategies are "decide at 0, at 1, or later"
# pOther = 【P_A0, P_A1, P_B0, P_C0, P_D0, P_E0】
# Calculate expected utility of action a given other players' mixing prob. (Turocy,for network)

def payoff(pOther, cost):
    '''This is for network (a)'''
    diff_pay = 100
    q = 0.7
    if cost=='h':
        c = 8
    elif cost == 'l':
        c = 1
    # We know there are 5 nodes, pay matrix is 5X3 
    pay =  np.empty((5,6),dtype=float)
    pay[:,0] = ProbCorrect_Bayesian(1,q)*diff_pay
    
    # Note: since here the payoff is about turn 0,1,2, need to do the max in the 
    # indegree = 2, node A
    pay_1 = ProbCorrect_Bayesian(3,q)*diff_pay - c
    pay_2 = pOther[2]*ProbCorrect_Bayesian(3,q)*diff_pay + (1- pOther[2])*ProbCorrect_Bayesian(5,q)*diff_pay - 2*c
    pay[0,1] = pay_1
    pay[0,2] = pay_2
    j = 2
    i = 1
    while j < 5:        
        j = j+ 1
        pay[0,j] = pay_2 - i*c
        i += 1
    
    # indegree = 3, node B
    pay_1 = ProbCorrect_Bayesian(4,q)*diff_pay - c
    pay_2 = pOther[0]*ProbCorrect_Bayesian(4,q)*diff_pay + (1- pOther[0])*ProbCorrect_Bayesian(5,q)*diff_pay - 2*c
    pay[1,1] = pay_1
    pay[1,2] = pay_2
    j = 2
    i = 1
    while j < 5:        
        j = j+ 1
        pay[1,j] = pay_2 - i*c
        i += 1
    
    
    # indegree = 1, node C or D, (position 2,3)
    pay_1 = ProbCorrect_Bayesian(2,q)*diff_pay - c
    pay_2 = pOther[0]*ProbCorrect_Bayesian(2,q)*diff_pay + (1-pOther[0])*ProbCorrect_Bayesian(4,q)*diff_pay - 2*c
    pay_3 = pOther[0]*ProbCorrect_Bayesian(2,q)*diff_pay + (pOther[1] + (1-pOther[0]-pOther[1])*pOther[2])*ProbCorrect_Bayesian(4,q)*diff_pay + (1-pOther[0]-pOther[1])*(1-pOther[2])*ProbCorrect_Bayesian(5,q)*diff_pay - 3*c
       
    pay[2,1] = pay_1
    pay[3,1] = pay_1
    pay[2,2] = pay_2
    pay[3,2] = pay_2
    pay[2,3] = pay_3
    pay[3,3] = pay_3
    j = 3
    i = 1
    while j < 5:        
        j = j+ 1
        pay[2,j] = pay_3 - i*c
        pay[3,j] = pay_3 - i*c
        i += 1
    
    
    # indegree = 1, node E, (position 4)
    pay_2 = pOther[0]*ProbCorrect_Bayesian(2,q)*diff_pay + (1- pOther[0])*ProbCorrect_Bayesian(3,q)*diff_pay - 2*c
    pay_3 = pOther[0]*ProbCorrect_Bayesian(2,q)*diff_pay + (pOther[1] + (1-pOther[0]-pOther[1])*pOther[2])*ProbCorrect_Bayesian(3,q)*diff_pay + (1-pOther[0]-pOther[1])*(1-pOther[2])*ProbCorrect_Bayesian(5,q)*diff_pay - 3*c

    pay[4,1] = pay_1
    pay[4,2] = pay_2
    pay[4,3] = pay_3
    j = 3
    i = 1
    while j < 5:        
        j = j+ 1
        pay[4,j] = pay_3 - i*c
        
        i += 1

    return pay
    
    
```

## Part 2: Compute QRE

Assumptions:
    * Five agents
    * Five nodes, the prob of deciding at turn 0
 #   <font color='red'> *  Next: prob of deciding at turn 1, given turn 0's outcome! </font>

```python

#Define the function whose 0 we want to find  (Turocy)
def F(p, l = 1,cost='h'): ## probability vector and value of lambda
    # The part where it is about prob of deciding at turn 0
  
    p_0s = np.delete(p, 1)
    p_0s_t =  np.transpose(p_0s)
    p_real = np.exp(p)
#     p_1s_t = np.log(1 - np.exp(p_0s_t))  
#     p_1s_t[0] = p[1]

#     pay_turn1 = payoff(p, cost)[:,1:].max(1)
#     pay_turn1[0] = payoff(p, cost)[0,1]
    
#     H= p_0s_t - p_1s_t - l*(payoff(p, cost)[:,0]- pay_turn1)
    
#     p_A2 = np.log(1 - np.exp(p[0]) - np.exp(p[1]))
   
#     A_1 = p[1] - p_A2 - l*(payoff(p,cost)[0,1] - np.max(payoff(p,cost)[0,2:]))


    # This version: the probability is the difference between prob of deciding at turn 0 vs later
    
    H= p_0s_t - np.log(1 - np.exp(p_0s_t))  - l*(payoff(p_real, cost)[:,0]- payoff(p_real, cost)[:,1:].max(1))
    # But additionally, we need subjects to expect the prob of A deciding at turn 1 or later
    p_A2 = np.log(1 - np.exp(p[0]) - np.exp(p[1]))
    A_12 =  p[1] - p_A2 - l*(payoff(p_real,cost)[0,1] - np.max(payoff(p_real,cost)[0,2:]))
    
    H_A = np.append(H, A_12)
    A_02 = p[0] - p_A2 - l*(payoff(p_real,cost)[0,0] - np.max(payoff(p_real,cost)[0,2:]))
    H_A = np.append(H_A, A_02)
    A_01 = p[0] - p[1] - l*(payoff(p_real,cost)[0,0] - payoff(p_real,cost)[0,1])
    H_A = np.append(H_A, A_01)
    
    res =  np.dot(np.transpose(H_A),H_A)
#     res =  np.dot(np.transpose(H),H) 
    return res

cons = ({'type': 'ineq', 'fun' : lambda p: 1- np.exp(p[0]) - np.exp(p[1])-0.00000001 } ,
       {'type': 'ineq', 'fun' : lambda p: 1- np.exp(p[2])-0.00000001 },
        {'type': 'ineq', 'fun' : lambda p: 1- np.exp(p[3])-0.00000001 },
        {'type': 'ineq', 'fun' : lambda p: 1- np.exp(p[4]) -0.00000001},
        {'type': 'ineq', 'fun' : lambda p: 1- np.exp(p[5])-0.00000001},
       )


cost='l'
lSeq = np.arange(0,10,.25)
QRE = np.zeros((len(lSeq),6))
p0 = np.log([0.25,0.25,0.5,0.5,0.5,0.5])
for (i,l) in enumerate(lSeq):
    res = minimize(F, p0, args=(l,cost,), method='SLSQP',constraints=cons)
    p0 = res.x # use the new value as the new guess for next lambda
    QRE[i] =  np.exp(res.x)
    
#Plot
fig, ax = plt.subplots()
ax.plot(lSeq, QRE[:,0],lSeq, QRE[:,1],lSeq, QRE[:,2],lSeq, QRE[:,3],lSeq, QRE[:,4],lSeq, QRE[:,5])
ax.set_xlabel('$\lambda$')
handles, labels = ax.get_legend_handles_labels()
lgd = ax.legend(handles, labels = ['A','A_1','B', 'C','D','E'], loc = 0)
ax.set_title("QRE when cost="+str(cost)+" Prob of decision at turn 0")
plt.show()    
```
