```python
def nCr(n,r):
    f = math.factorial
    return f(n) / f(r) / f(n-r)
```

```python
def ProbCorrect_Bayesian(n, q):
    '''Based on Bayesian rule, the prob of making a correct guess by observing n signals '''
    nn = 0
    prob = 0
    
    while nn <=  (n/2):
        
        #print(nn)
        #if ((n-nn) % 2 ) == 0 & (nn!=0) :
        if nn ==  (n/2) :
            prob += nCr(n,(n-nn))*(q**(n-nn))*((1-q)**(nn))*(1/2)
            #print(prob)
        else :
            prob += nCr(n,(n-nn))*(q**(n-nn))*((1-q)**(nn))
            #print(prob)
        nn += 1
        
    
    return prob
    
    
```

```python
ProbCorrect_Bayesian(3,0.7) - (ProbCorrect_Bayesian(5,0.7)-ProbCorrect_Bayesian(1,0.7))/4*3
```

```python
def ProbLinear(n,q):
    # 1. the difference remain the same, ProbCorrect_Bayesian(5,q)-ProbCorrect_Bayesian(1,q) 
    # When q = 0.7, the difference is (ProbCorrect_Bayesian(5,0.7)-ProbCorrect_Bayesian(1,0.7))/4 = 0.3423
    # 2. Ensure that the middle point is the correct probability
    # ProbCorrect_Bayesian(3,0.7) = a + 0.3423*3
    # a = 0.68131
    
    b = ProbCorrect_Bayesian(5,q)-ProbCorrect_Bayesian(1,q))/4
    a = ProbCorrect_Bayesian(3,q) - b*3
    
    return a+b*n
    
    
```

```python
# Define the agent
class Agent: 
    ''' Class definition of the Agent in network model'''
    
    def __init__(self, location, InforSet, waiting_cost, payoff_correct, payoff_wrong, q, period, network,  action=-1,):
        '''Initialize agent.
            location -- location of the agent, index of agent in the environment
            
            InforSet -- the information set this agent observes, includes the index number of the observed signals,
                        {1,2,3,...}, takes the form of []
            
            waiting_cost -- waiting cost for waiting for one more period
            payoff_correct -- the payoff amount for a correct guess
            payoff_wrong -- the payoff amount for a wrong guess
            q: informativeness
            
            period: the period played in the network
            network -- the network this agent is in
            
            action -- action of the agent, the time that agent decides to guess, t is in {0,1,2,3...},
                        when action is -1, not decided yet
            
            '''
        
        
        #infor_set -- the number of unique signals that the agent can observe at time t
        
        self.location = location
        self.InforSet = InforSet
        
        self.waiting_cost = waiting_cost
        self.payoff_correct = payoff_correct
        self.payoff_wrong = payoff_wrong
        self.q = q
        
        self.period = period
        self.network = network
        
        self.action = action   
     
    def StopWhenInDegree1(self):
        if self.network.in_degree(self.location) == 1 : 
            self.action = self.period
        
        return self.action
    
    def MaxWaitingPeriod(self):
            # The max waiting period is jointly determined by the waiting cost and the expected number of additional private signals that can be observed in each period 
            # Compare 2 numbers:
            # Based on the cost, the max number of peirods can afford
            # Based on the connectiveness, the longest path 
        
        # Firstly, based on the wating cost and the payoff schemes to decide the max waiting period
        if self.waiting_cost == 0 :
            EconMaxWaiting =  math.inf
        else :
            EconMaxWaiting = int((self.payoff_correct- self.payoff_wrong) / self.waiting_cost)
            
        self.MaxPeriod = np.zeros(nx.number_of_nodes(self.network))
        for i in range(nx.number_of_nodes(self.network)):    
            #print(i)
            try:
                paths = list(nx.shortest_simple_paths(self.network,  i+1, self.location))
            #print(paths)
                length = [len(x) for x in paths]
                self.MaxPeriod[i]=min(max(length)-1, EconMaxWaiting)
            except nx.NetworkXNoPath:
                self.MaxPeriod[i]=EconMaxWaiting
            
            
                
        return self.MaxPeriod
    
    def MinWaitingPeriod(self):
            # The shortest path that this agent can observe the other agent's private signal 
            
        self.MinPeriod = np.zeros(nx.number_of_nodes(self.network))
        for i in range(nx.number_of_nodes(self.network)):    
            #print(i)
            try:
                paths = list(nx.shortest_simple_paths(self.network,  i+1, self.location))
            #print(paths)
                length = [len(x) for x in paths]
                self.MinPeriod[i]=min(length)-1
            except nx.NetworkXNoPath:
                self.MinPeriod[i]=math.inf
                
        return self.MinPeriod
        
    def OptimalStoppingRule(self):
        # Optimal rule depends only on the MinPeriod, the best case scenario
        # Step 1: Use the self.MinPeriod to decide how many additional private signals can be observed 
        # Step 2: Compare for each period, the added expected payoff vs the incurred additional cost
        ExpectedPayoffs = []
        
        
#         MaxPeriod = 0
#         MaxExpectedPayoff = 0
#         AddedPayoff = 0 
#         IncurredCost = 0
        
#         if IncurredCost<= AddedPayoff:
#             Period += 1
#             NumSignals = sum(self.MinPeriod<= Period)
#             NewExpectedPayoff = self.payoff_wrong +  (self.payoff_correct  - self.payoff_wrong) * ProbCorrect_Bayesian(NumSignals, self.q)
#             AddedPayoff = NewExpectedPayoff - ExpectedPayoff
#             IncurredCost = self.waiting_cost
            
#       for Period in range(self.period, 11) :
        inforset = copy.deepcopy(self.InforSet)
        for Period in range(self.period, 11) :
            for j in range(nx.number_of_nodes(self.network)) :
                #if (( self.MinPeriod[j] + self.period ) <= Period)  and (j+1 not in self.InforSet) :
                if ( self.MinPeriod[j]  <= Period)  and (j+1 not in inforset) :
                    inforset.append(j+1)

                    
            NumSignals = len(inforset)
            ExpectedPayoff = self.payoff_wrong +  (self.payoff_correct  - self.payoff_wrong) * ProbCorrect_Bayesian(NumSignals, self.q) -  (self.waiting_cost*Period)
            ExpectedPayoffs.append((Period,NumSignals, ExpectedPayoff,inforset ))
            
            
        self.ExpectedPayoffs_data = pd.DataFrame(columns = ['Period', 'NumSignals' ,'ExpectedPayoff', "InforSet"], data = ExpectedPayoffs)
        
         
        optimal = self.ExpectedPayoffs_data['ExpectedPayoff'].idxmax()
        
        # Before, tried to use the period, not working
        # Instead, use the number of signals as the stopping rule 
        return self.ExpectedPayoffs_data.loc[optimal, :]

        
#         return "Period: "+str(Period)+" ; Number of Signals:"+str(NumSignals)
            

        
        
     #   def MaxExpectedPayoff(self):
            # Based on the network position, an agent should be able to calculate her max expected payoff
            # which is jointly decided by the agent's own position in the network, the knowledge of the network, and other participants' decisions
            # It can be reached when the number of one signal reaches a certain amount/ the period has reached a certain time 
            # The first case means that signal draws might be better than expected, so the decision will happen earlier
            # The second case means that one the agent reaches this period, there won't be more information coming in, should make a decision (last resort)
            # If they use the bayesian updating probability to calculate their expected payoff, then the optimal stopping rule will be a threshold of number of private signals this agent can observe 
            # Here it is either the number of private signals or the number of periods
            
    def __str__(self):
        return "Agent!"
```

```python
# Define the environment
class DirectedNetwork: 
    ''' Class definition of the Environment'''
        
    def __init__(self, Network, TrueNature, Sig_Acc,  Waiting_Cost, Payoff_Correct, Payoff_Wrong, Period = 0, Decisions=[], InforSets=[]):
        
        
        '''Initialize environment.
            Network -- network representation
            TrueNature -- the true nature status, 1 or -1
            Sig_Acc -- signal informativeness
            Period -- timing of the game
            Decisions -- [0,1,...] the decisions of all participants, the Period that this agent makes a guess 
            InforSets -- the inforset each participant has at the begining of each Period 
        '''
        self.Network = Network
        self.TrueNature = TrueNature
        self.Sig_Acc = Sig_Acc
        self.Waiting_Cost = Waiting_Cost
        self.Payoff_Correct = Payoff_Correct
        self.Payoff_Wrong = Payoff_Wrong
        
        self.Period = Period
        if Decisions == [] :
            self.Decisions = -1 * np.ones(self.Network.number_of_nodes())
        else :
            self.Decisions = Decisions
        
        if InforSets == [] and (self.Period == 0) :
            self.InforSets = [ [x] for x in range(1, 1+self.Network.number_of_nodes())]
        else :
            self.InforSets = InforSets
            
            
        self.InforFlow = nx.adjacency_matrix(self.Network)
        
    def AssignSignals(self):
        self.SignalDraw = np.random.rand(self.Network.number_of_nodes())
        self.Signals = np.zeros(self.Network.number_of_nodes())
        for i in range(self.Network.number_of_nodes()):
            if (self.SignalDraw[i] <= self.Sig_Acc):
                self.Signals[i] = self.TrueNature
            else :
                self.Signals[i] = -1 *self.TrueNature
        
        return self.Signals
        
    
    def SimultaneousMove(self) :
        print('this is Period '+str(self.Period))
        # Obtain the decision for each agent given the current Period inforset
        for j in range(self.Network.number_of_nodes()):
            if self.Decisions[j] == -1 :
                agent = Agent( j+1 , self.InforSets[j], self.Waiting_Cost, self.Payoff_Correct, self.Payoff_Wrong, self.Sig_Acc, self.Period, self.Network)
    #             print(str(j+1)+str([self.InforSets[j]]))
                agent.MinWaitingPeriod()
                agent.OptimalStoppingRule()
               # print(agent.InforSet)
                print('Agent '+str(j))
                print(agent.ExpectedPayoffs_data)
                print('Highest Payoff'+str(agent.ExpectedPayoffs_data['ExpectedPayoff'].idxmax()))
                print(agent.OptimalStoppingRule())

                optimal = agent.ExpectedPayoffs_data['ExpectedPayoff'].idxmax()
                self.OptimalNumSig = agent.ExpectedPayoffs_data['NumSignals'].loc[optimal] 
            
               # print('Agent '+str(j+1)+'s Optimal Number of Signals'+str(self.OptimalNumSig ))
                
                
                
                # Note: here if the agent makes the decision, then the network reduces!
                #if self.Period in self.OptimalPeriod :
               # print('Inforset'+str(self.InforSets))
                if  self.OptimalNumSig == len(self.InforSets[j]) :
                    self.Decisions[j] =self.Period
    
    def SimultaneousMove_linear(self) :
        print('this is Period '+str(self.Period))
        # Obtain the decision for each agent given the current Period inforset
        for j in range(self.Network.number_of_nodes()):
            if self.Decisions[j] == -1 :
                agent = Agent_linear( j+1 , self.InforSets[j], self.Waiting_Cost, self.Payoff_Correct, self.Payoff_Wrong, self.Sig_Acc, self.Period, self.Network)
    #             print(str(j+1)+str([self.InforSets[j]]))
                agent.MinWaitingPeriod()
                agent.OptimalStoppingRule()
               # print(agent.InforSet)
                print('Agent '+str(j))
                print(agent.ExpectedPayoffs_data)
                print('Highest Payoff'+str(agent.ExpectedPayoffs_data['ExpectedPayoff'].idxmax()))
                print(agent.OptimalStoppingRule())

                optimal = agent.ExpectedPayoffs_data['ExpectedPayoff'].idxmax()
                self.OptimalNumSig = agent.ExpectedPayoffs_data['NumSignals'].loc[optimal] 
            
               # print('Agent '+str(j+1)+'s Optimal Number of Signals'+str(self.OptimalNumSig ))
                
                
                
                # Note: here if the agent makes the decision, then the network reduces!
                #if self.Period in self.OptimalPeriod :
               # print('Inforset'+str(self.InforSets))
                if  self.OptimalNumSig == len(self.InforSets[j]) :
                    self.Decisions[j] =self.Period

        
    def NextPeriod(self):
        #Two steps: 1. update the inforset, regardless of each agent's decision, the infor will be transimitted using the previous network stucture
        # 2. based on the decisions, the network is "down-graded"
        
        # Firstly, define the inforset 
        # Here the NextInforSets is updated using current network structure
        self.NextInforsets = copy.deepcopy(self.InforSets)
        #self.NextInforsets = []
        for j in range(self.Network.number_of_nodes()) :
            #self.NextInforsets[j] = copy.deepcopy(self.InforSets[j])
            self.temp = copy.deepcopy(self.InforSets[j])
            for i in range(self.Network.number_of_nodes()):
                #if (i != j) and self.Network.has_edge(i+1,j+1) and (i+1 not in self.temp):
                if (i != j) and self.Network.has_edge(i+1,j+1) :
                    for x in self.InforSets[i] :
                        if x not in self.temp:
                            self.temp.append(x)         
            self.NextInforsets[j]=self.temp
            #print("InforSets: "+str(self.InforSets))
            #print('Next Inforsets: '+str(self.NextInforsets))
      
        #return self.NextInforsets
        
        # Secondly, based on decisions, "down-grade" the network structure
        self.NextNetwork = copy.deepcopy(self.Network)
        for j in range(self.Network.number_of_nodes()) :
            if  self.Decisions[j] == self.Period :
                self.NextNetwork.remove_edges_from(list(self.NextNetwork.in_edges(j+1)))
        
        #return self.NextNetwork            
        
        # Thirdly, update the next Period number
        self.NextPeriod = self.Period + 1
              
        
    
                

     
            

        
    def __str__(self):
        plt.title("True Nature is: "+str(self.TrueNature)+"; Signal Informativeness: "+str(self.Sig_Acc))
        nx.draw(self.Network, pos=nx.circular_layout(self.Network), with_labels=True, node_size=1500, alpha=1, arrows=True) 
       # plt.show()
        return "True Nature is: "+str(self.TrueNature)+"; Signal Informativeness: "+str(self.Sig_Acc)
    
```

```python
# Define an environment to calculate Behavioral Prediction, additional input as the realized signals 
class DirectedNetworkAndRealizedSignals:

    ''' Class definition of the Environment that helps calculate the behavioral prediction:
            this is for a naive agent who has the following decision rule:
            1. decide whether he should wait or not at p=0, based on whether he can observe additional signal or not
            2. if he can observe additional signals, always wait to period 1
            3. if the signals are not informative, then wait for another signal
            4. Stop until the agent can make a decision or the information set does not increase in the next period
    '''

    def __init__(self, DirectedNetwork, Signals, Half_Guess=[], Self_Guess=[] ):
        
        
        '''Initialize environment.
            Network -- the orignals directed network environment
            Signals -- the realized signals
        '''
        self.DirectedNetwork = copy.deepcopy(DirectedNetwork)
        self.Signals = copy.deepcopy(Signals)
        
        self.Period = self.DirectedNetwork.Period
        
        if (Half_Guess==[]) and (self.Period == 0) :
            self.Half_Guess = np.zeros(nx.number_of_nodes(self.DirectedNetwork.Network))
        else :
            self.Half_Guess = Half_Guess
        
        if (Self_Guess ==[]) and (self.Period == 0) :
            self.Self_Guess = np.zeros(nx.number_of_nodes(self.DirectedNetwork.Network))
        else :
            self.Self_Guess = Self_Guess
        
#         if InforSets == [] and (self.Period == 0) :
#             self.InforSets = [ [x] for x in range(1, 1+self.DirectedNetwork.Network.number_of_nodes())]
#         else :
#             self.InforSets = InforSets
            
        self.Decisions = self.DirectedNetwork.Decisions
        self.InforSets = self.DirectedNetwork.InforSets
        
        
    def SimultaneousMove(self) :
        print('This is Period '+str(self.Period))
        self.NewInforsets = copy.deepcopy(self.InforSets)
        # Obtain the decision for each agent given the current Period inforset
        for j in range(nx.number_of_nodes(self.DirectedNetwork.Network)):
           
            if self.Decisions[j] == -1 :
                agent = Agent( j+1 , self.InforSets[j], self.DirectedNetwork.Waiting_Cost, 
                              self.DirectedNetwork.Payoff_Correct, self.DirectedNetwork.Payoff_Wrong, 
                              self.DirectedNetwork.Sig_Acc, self.Period, self.DirectedNetwork.Network)
            # Whenever there is no incoming links, stop
                if agent.StopWhenInDegree1() != -1 :
                    self.Decisions[j] = agent.StopWhenInDegree1()
                    self.Half_Guess[j] = self.Signals[j]
                    self.Self_Guess[j] = self.Signals[j]
            
#             # Obtain the "Expected" new (next) period inforsets
#             if self.Decisions[j] == -1 :
#                 temp = self.InforSets[j]
#                 for i in range(nx.number_of_nodes(self.DirectedNetwork.Network)):
#                     if (i != j) and self.DirectedNetwork.Network.has_edge(i+1,j+1) :
#                         for x in self.InforSets[i] :
#                             if x not in temp:
#                                 temp.append(x)         
#                 self.NewInforsets[j]=temp
   
            if self.Decisions[j] == -1 and self.Period >0 :
#                 print("Run to Here")
                # Based on the signals' realization, makes a decision        
                Decide = 0
                for x in  self.InforSets[j] :
                    Decide += self.Signals[x-1]
                
#                 print('Decide')
#                 print(Decide)
                    
                if Decide > 0 :
                        self.Decisions[j] = self.Period
                        self.Half_Guess[j] = 1
                        self.Self_Guess[j] = 1
                elif  Decide < 0 :
                        self.Decisions[j] = self.Period
                        self.Half_Guess[j] = -1
                        self.Self_Guess[j] = -1
#                 elif Decide == 0 and self.NewInforsets[j]==self.DirectedNetwork.Inforsets[j] :
                elif Decide == 0 :
        # Note here the next period signal sets is the "expected" infor sets based on current network, 
        # may not be the real next period network 
        # For any signal that is not in the inforset, if the min_period to obtain is to wait for infinite time (no way to pass through), then decide
                    agent.MinWaitingPeriod()
                    for i in range(nx.number_of_nodes(self.DirectedNetwork.Network)):    
                        if (i not in self.InforSets[j] ) and agent.MinPeriod[i]==math.inf:
                            self.Decisions[j] = self.Period 
                            self.Half_Guess[j] = 2*np.random.randint(2)-1
                            self.Self_Guess[j] = self.Signals[j]

        return print(self.Self_Guess)

        
    def NextPeriod(self):
        #Two steps: 1. update the inforset, regardless of each agent's decision, the infor will be transimitted using the previous network stucture
        # 2. based on the decisions, the network is "down-graded"
        
        # Firstly, define the inforset 
        # Here the NextInforSets is updated using current network structure
        self.NextInforsets = copy.deepcopy(self.InforSets)
        #self.NextInforsets = []
        for j in range(nx.number_of_nodes(self.DirectedNetwork.Network)) :
            #self.NextInforsets[j] = copy.deepcopy(self.InforSets[j])
            self.temp = copy.deepcopy(self.InforSets[j])
            for i in range(nx.number_of_nodes(self.DirectedNetwork.Network)):
                #if (i != j) and self.DirectedNetwork.Network.has_edge(i+1,j+1) and (i+1 not in self.temp):
                if (i != j) and self.DirectedNetwork.Network.has_edge(i+1,j+1) :
                    for x in self.InforSets[i] :
                        if x not in self.temp:
                            self.temp.append(x)         
            self.NextInforsets[j]=self.temp
            #print("InforSets: "+str(self.InforSets))
            #print('Next Inforsets: '+str(self.NextInforsets))
      
        #return self.NextInforsets
        
        # Secondly, based on decisions, "down-grade" the network structure
        self.NextNetwork = copy.deepcopy(self.DirectedNetwork.Network)
        for j in range(nx.number_of_nodes(self.DirectedNetwork.Network)) :
            if  self.Decisions[j] == self.Period :
                self.NextNetwork.remove_edges_from(list(self.NextNetwork.in_edges(j+1)))
        
        #return self.NextNetwork            
        
        # Thirdly, update the next Period number
        self.NextPeriod = self.Period + 1
              
        
    def __str__(self):

        return "The directed network environment is : "+str(self.DirectedNetwork)+"; Realized Signals are : "+str(self.Signals)

```
