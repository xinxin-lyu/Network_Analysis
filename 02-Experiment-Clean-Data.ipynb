```python
columns = {'PDCN': ['.player.robot_signal', '.player.robot_type', '.player.robot_period','.player.robot_guess','.subsession.storyline'],
          'DCN': [ '.player.sec_spent', '.group.storyline', '.subsession.groupnumber', '.group.id_in_subsession']
          }

shared_columns = ['.player.current_ball', '.player.type','.player.period', '.player.choice_of_urn',  
                 '.player.decision_period','.player.information_set_lastperiod', '.player.information_set_currentperiod',
                  '.player.history','.player.time_on_page',
                 '.subsession.round_number']

additional_columns = ['participant.code','session.code','network','waiting_cost','date','app_name']

ncols = 40
```

```python
# for j,part_name in enumerate(['part2_play.','part3_play.']):
result = {}
for part_name in columns.keys() :
    out = {}
    ncols = 40
    for col in additional_columns+shared_columns+columns[part_name]:
        out[ col]=np.empty((nRows,ncols))
        out[ col][:]=np.nan

    for i,s in enumerate(sessions):

        print(s)
        data=pd.read_csv(datafolder+'\\'+s)
        start_row = sessions[s][4]
        end_row = sessions[s][5]
    #     for j,part_name in enumerate(['PDCN', 'DCN']):

        if part_name == 'PDCN':
            ncols = 10
        else :
            if i == 0:
                ncols = 40
            else :
                ncols = 30
        
        for col in columns[part_name]+shared_columns:
            if i <6 :
                colnames = [part_name+'.'+str(x)+col for x in range(1,ncols+1)]
            else :
                colnames = [part_name+'_b.'+str(x)+col for x in range(1,ncols+1)]
            temp=data[colnames].to_numpy()
            if type(temp[0][0])==str:
                out[ col]=out[ col].astype(str)
            out[ col][start_row:end_row,0:ncols]=temp

            #sessionID
            colnames = ['session.code' for x in range(1,ncols+1)]
            temp=data[colnames].to_numpy()
            if type(temp[0][0])==str:
                out['session.code']=out['session.code'].astype(str)
            out[ 'session.code'][start_row:end_row,0:ncols]=temp

            #subjectID
            colnames = ['participant.code' for x in range(1,ncols+1)]
            temp=data[colnames].to_numpy()
            if type(temp[0][0])==str:
                out[ 'participant.code']=out['participant.code'].astype(str)
            out[ 'participant.code'][start_row:end_row,0:ncols]=temp

            #treatment:


            network = sessions[s][2]
            out['network']=out['network'].astype(str)
            out['network'][start_row:end_row,0:ncols]=network

            waiting_cost = sessions[s][3]
            out['waiting_cost']=out['waiting_cost'].astype(str)
            out['waiting_cost'][start_row:end_row,0:ncols]=waiting_cost

            date = sessions[s][1]
            out['date']=out['date'].astype(str)
            out['date'][start_row:end_row,0:ncols]=date

            app_name = part_name
            out['app_name']=out['app_name'].astype(str)
            out['app_name'][start_row:end_row,0:ncols]=app_name


#         temp = sessions[s][0]
#         if type(temp)==str:
#             out['session']=out['session'].astype(str)
#         out['session'][start_row:end_row,0:ncols]=temp
#     if result == {} :
#         result.update(out)
#     else :
#         result = combine_dict(result, out)
#     print(result['app_name'])
    for col in out:
        out[col]=out[col].flatten()
    result[part_name] = out
```

```python
out1 =  pd.DataFrame(result['PDCN'])
out2 =  pd.DataFrame(result['DCN'])
```

```python
out = pd.concat([out1, out2])
out = out[out['app_name']!='nan']
out.to_csv("02-Experiment-Clean-Data.csv",index=False)
```
