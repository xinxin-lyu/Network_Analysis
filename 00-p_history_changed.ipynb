```python
datafolder = "C:\\Users\\xinxi\\Purdue\\EffectiveSocialLearningProject\\DataAnalysis\\ImportedData\\"
```

```python
 data = pd.read_csv(datafolder+"p_history.csv", sep=",", header=0)
```

```python
data['last_infor_content'] = ""
data['num_red'] = 0
data['num_blue'] = 0
```

```python
for i in range(len(data['p_history'])):
    
    data['last_infor_content'][i] =  data['p_history'][i].split("[[")[1].split("{")[-1].split("},")[0]
    red = 0
    blue = 0
    for x in data['last_infor_content'][i].split(":"):
        if "red" in x :
            red += 1
        if "blue" in x :
            blue += 1
    data['num_red'][i] = red
    data['num_blue'][i] = blue
        
            
```

```python
data.to_csv(datafolder+"p_history_changed.csv", sep=",")
```
